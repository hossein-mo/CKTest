```python
#If you want to use a cuda device make sure to configure it correctly.
from cktest import CKTest as ckt
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.normal(size=100000)
ck = ckt(x, dx=1, bins_mode='std', bins=0)
```

```python
ck = ckt(x, dx=1, bins_mode='std', bins=0)
print('PDF shape = {}'.format(ck.pdf_shape))
plt.figure(figsize=(16,9), dpi=60)
plt.plot(x,',')
plt.hlines(ck.xbins, 0, len(x), zorder=10)
plt.show()
```

```python
cktest = ck.run_test(2000)
ck.plot()
```
